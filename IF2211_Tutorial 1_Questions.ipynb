## **1. NumPy Basics**

### **Q1: Array Creation & Operations**
- Create a NumPy array with values from 1 to 20.
- Reshape it into a 4×5 matrix.
- Find the mean, median, and standard deviation.
- Extract all even numbers from the array.

**Guidance:** Use `np.arange()` to create the array, `reshape()` to change its shape, and functions like `np.mean()` for calculations.

```python
# Write your code here
import numpy as np

array1 = np.arange(1, 21)

array1.reshape(4, 5)

mean, median, std = np.mean(array1), np.median(array1), np.std(array1)

mean, median, std

even_numbers = array1[array1 % 2 == 0]
even_numbers
```

### **Q2: Broadcasting & Indexing**
- Create a 5×5 matrix with random numbers.
- Replace all values greater than 0.5 with `1`, and others with `0`.
- Extract the third column of the matrix.

**Guidance:** Use `np.random.rand()` for generating random values and boolean indexing for modifications.

```python
# Write your code here
np.random.seed(42)
array2 = np.array(np.random.rand(5, 5))
array2

array2[array2 > 0.5 ] = 1
array2[array2 <= 0.5] = 0
array2

array2[:, 2].astype(int)
```

## **2. pandas for Data Handling**

### **Q3: Basic DataFrame Operations**
- Create a DataFrame with three columns: `Name`, `Age`, and `Score`.
- Add a new column `Passed` where `Score > 50` is `True`, otherwise `False`.
- Drop rows where `Age` is missing.

**Guidance:** Use `pd.DataFrame()` to create the table, and functions like `dropna()` for cleaning.

```python
# Write your code here
import pandas as pd

df = pd.DataFrame({
    'Name': ['John', 'Jane', 'David'],
    'Age': [25, 30, 35],
    'Score': [30, 90, 75]
})

df['Passed'] = df['Score'] >= 60
df

df.dropna()
```

### **Q4: Data Cleaning & Aggregation**
- Load a dataset (`titanic.csv` or another sample dataset).
- Drop duplicate rows.
- Fill missing values in the `Age` column with the median.
- Group by `Sex` and calculate the mean `Age`.

**Guidance:** Use `pd.read_csv()` to load the dataset and `groupby()` for aggregations.

```python
# Write your code here
titanic_df = pd.read_csv('titanic.csv')

titanic_df = titanic_df.drop_duplicates().dropna()
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df = titanic_df.groupby('Sex')['Age'].mean()
titanic_df
```

## **3. Data Visualization**

### **Q5: Matplotlib Basics**
- Plot a line graph showing the relationship between `x = np.arange(0, 10, 0.1)` and `y = np.sin(x)`.
- Label the axes and add a title.

**Guidance:** Use `plt.plot()` from Matplotlib to create the graph, and `xlabel()`/`ylabel()` to label axes.

```python
# Write your code here
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sinusoidal Function')
plt.legend()
plt.show()
```

### **Q6: pandas & Seaborn**
- Load a dataset (e.g., `iris` from `seaborn`).
- Plot a histogram of the `sepal_length` column.
- Create a scatter plot of `sepal_length` vs. `sepal_width`, colored by species.

**Guidance:** Use Seaborn's `sns.histplot()` and `sns.scatterplot()` for visualization.

```python
# Write your code here
import seaborn as sns

iris = pd.read_csv('iris.csv')
sns.histplot(data=iris, x='sepal_length', bins=10, kde=True)
plt.show()

sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```
